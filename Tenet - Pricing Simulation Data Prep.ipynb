```python
import spss, spssdata

RevHas = 2743
ProcHas = 'J1557Z'
AnyDiagHas = 'A299.0'
PrimDiagHas = 'A299.0'

simDataFile = "C:/Projects/Tenet/temp/Showcase prepped v3""" >< ^nodeBaseName >< """.sav"
simXML = "C:/Projects/Tenet/temp/simPrepModel.xml"
simPlanFile = "C:/Projects/Tenet/temp/currentSimPlan.splan"
simOutputFile = "C:/Projects/Tenet/temp/SimPOC3_v1_""" >< ^nodeBaseName >< """.sav"

getPayer_chargecode = '\n'.join([
'GET DATA /TYPE=XLSX',
'  /FILE=\'C:/Projects/Tenet/temp/payer_chargecode2""" >< ^nodeBaseName >< """.xlsx\' ',
'  /SHEET=name \'Sheet1\' ',
'  /CELLRANGE=full ',
'  /READNAMES=on ',
'  /ASSUMEDSTRWIDTH=32767.',
'CACHE.',
'EXECUTE.',
'DATASET NAME BCBSdata WINDOW=FRONT.',
])
spss.Submit(getPayer_chargecode)

###########################################
#
#  SimPrep: generate the SPSS Syntax for the SimPrep command
#
###########################################
simprep = '\n'.join([
'*Simulation preparation. ',
'SIMPREP BEGIN ',
'/OUTFILE PREPXML=\'' + simXML + '\'.',
'NUMERIC \n',
])
data = spssdata.Spssdata()
for row in data:
	varName = row.Payer.replace(" ","") + row.chargecode.replace(" ","")
	#simprep += "  " + varName + " \n"
	simprep += "  " + varName + "_Price \n"
	simprep += "  " + varName + "_LOS \n"
	simprep += "  " + varName + "_GrossRev \n"
	simprep += "  " + varName + "_SL \n"
	simprep += "  " + varName + "_NetRev \n"
simprep += "  " + "TotalRev. \n"
data.CClose()

simprep += "VARIABLE LEVEL \n"
data = spssdata.Spssdata()
for row in data:
	varName = row.Payer.replace(" ","") + row.chargecode.replace(" ","")
	simprep += "  " + varName + "_LOS \n"
simprep += "  " + "(ORDINAL). \n"
data.CClose()

simprep += "VARIABLE LEVEL \n"
data = spssdata.Spssdata()
for row in data:
	varName = row.Payer.replace(" ","") + row.chargecode.replace(" ","")
	#simprep += "  " + varName + " \n"
	simprep += "  " + varName + "_Price \n"
	simprep += "  " + varName + "_GrossRev \n"
	simprep += "  " + varName + "_SL \n"
	simprep += "  " + varName + "_NetRev \n"
	simprep += "  " + varName + "_VarCost \n"
	simprep += "  " + varName + "_Margin \n"
simprep += "  " + "TotalRev (SCALE)."
data.CClose()

simprep += "\n\n"

##########################################################
# Payer Logic Setup and Initialization
##########################################################
import csv
# BCBS_PPO
execfile('C:\\Projects\\Tenet\\Contract Terms\\bcbs_ppo.py')


simprep += "COMPUTE TotalRev = "

data = spssdata.Spssdata()
i = 1
for row in data:
	varName = row.Payer.replace(" ","") + row.chargecode.replace(" ","")
	if i == 1:
		simprep += ""
	else:
		simprep += " + "
	simprep += varName + "_NetRev"
	i = i + 1
simprep += ".\n\n"
data.CClose()

simprep += "SIMPREP END.\n\n"

print simprep
spss.Submit(simprep)

###########################################
#
#  SimPlan: generate the SPSS Syntax for the SimPlan command
#
###########################################
simplan = '\n'.join([
'*Create simulation plan. ',
'SIMPLAN CREATE ',
' /MODEL FILE=\"' + simXML + '\"  CUSTOM=YES', 
' /CONTINGENCY MULTIWAY=YES ',
' /TARGETOPTS  TARGET=TotalRev(FORMAT=F,2) OUTPUT=YES',
])

data = spssdata.Spssdata()
for row in data:
	varName = row.Payer.replace(" ","") + row.chargecode.replace(" ","")
#	simplan += " " + "/SIMINPUT INPUT=" + varName + "(FORMAT=F,2 MAPTO=" + varName + ") OUTPUT=YES TYPE=AUTOFIT(CATEGORICAL) \n"
	simplan += " " + "/SIMINPUT INPUT=" + varName + "_Price(FORMAT=F,2) OUTPUT=YES TYPE=MANUAL(LOCK=NO) DISTRIBUTION=UNIFORM(MIN=" + str(row.Price*0.5) + " MAX=" + str(row.Price*1.5) + " ) \n"
	simplan += " " + "/SIMINPUT INPUT=" + varName + "_LOS(FORMAT=F,2 MAPTO=" + varName + "LOS) OUTPUT=YES TYPE=MANUAL(LOCK=NO) DISTRIBUTION=EMPIRICAL \n"
	simplan += " " + "/SIMINPUT INPUT=" + varName + "_VarCost(FORMAT=F,2 MAPTO=" + varName + "VarCost) OUTPUT=YES TYPE=MANUAL(LOCK=NO) DISTRIBUTION=EMPIRICAL \n"
data.CClose()	

simplan += '\n'.join([
' /AUTOFIT NCASES=200000 FIT=AD BINS=100 ',
' /STOPCRITERIA MAXCASES=1000',
' /MISSING CLASSMISSING=EXCLUDE ',
' /PLAN FILE=\'' + simPlanFile + '\' DISPLAY=YES.',
'',
])

getSimData = '\n'.join([
'GET ',
'  FILE=\"' + simDataFile + '\".',
'DATASET NAME simData WINDOW=FRONT.',
'',
])

print simplan	
spss.Submit(getSimData)
spss.Submit(simplan)

simrun = '\n'.join([
'*Run simulation plan. ',
'SIMRUN ',
' /PLAN FILE=\'' + simPlanFile + '\' ',
' /CRITERIA  REPRESULTS=TRUE  SEED=9999999 ',
' /DISTRIBUTION DISPLAY=PDF SCALE=PDFVIEW(CURVE) MEAN OVERLAYTARGETS(NO) CATEGORICAL= PREDVAL GROUP(CATS) ',
#' /BOXPLOT DISPLAY=NO ',
#' /SCATTERPLOT DISPLAY=NO ',
#' /TORNADO DISPLAY=YES TYPE=CORR VARIANCE DELTASTDDEV(0.5)',
' /PRINT ASSOCIATIONS=YES DESCRIPTIVES=YES(CILEVEL=95.0) PERCENTILES=NO ',
' /OUTFILE FILE=\'' + simOutputFile + '\'.',
'',
])

spss.Submit(simrun)
```
