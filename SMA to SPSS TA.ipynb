```python
#parse the JSON file and iterate over the types and concepts and create text files (*.txt) of all of them.

import sys, traceback, os
import json, csv, string
import unicodedata

validFilenameChars = "-_.() %s%s" % (string.ascii_letters, string.digits)

def removeDisallowedFilenameChars(filename):
    cleanedFilename = unicodedata.normalize('NFKD', filename).encode('ASCII', 'ignore')
    return ''.join(c for c in cleanedFilename if c in validFilenameChars)



ifile = "C:\\Users\\IBM_ADMIN\\Documents\\1 - Clients\\HTC\\pro00006_configuration.json"
ofile = "C:\\Users\\IBM_ADMIN\\Documents\\1 - Clients\\HTC\\pro00046_configuration (2).json"
wd = "C:\\Users\\IBM_ADMIN\\Documents\\1 - Clients\\HTC\\"
                        
#replace "\'" with "'" in order to clean json                        
with open(ofile, "wt") as fout:
    with open(ifile, "rt") as fin:
        for line in fin:
            fout.write(line.replace(r"\'", "'"))

with open(ofile) as jsoninput:
    #load the JSON file into a dict
    ontology = json.load(jsoninput)
os.remove(ofile)

#this will throw an error if the folder already exists
try:
    os.mkdir("C:\\Users\\IBM_ADMIN\\Documents\\1 - Clients\\HTC\\sma_for_import")
except:
    pass

for i in range(0,len(ontology['admin-app-config']['analysis-config']['ModelConfiguration'])): #theme
    for j in range(0,len(ontology['admin-app-config']['analysis-config']['ModelConfiguration'][i]['brands'])): #concept
        for k in ontology['admin-app-config']['analysis-config']['ModelConfiguration'][i]['brands'][j]['regexes'][0].keys(): #term group
            theme_name = ontology['admin-app-config']['analysis-config']['ModelConfiguration'][i]['name']
            concept_name = ontology['admin-app-config']['analysis-config']['ModelConfiguration'][i]['brands'][j]['name']
            group_name = k
            terms = ontology['admin-app-config']['analysis-config']['ModelConfiguration'][i]['brands'][j]['regexes'][0][k]
            delimiter = " - "
            #need to make new filename safe for file system (remove forbidden characters)
            newfile = theme_name+delimiter+concept_name+delimiter+group_name+".txt"
            print removeDisallowedFilenameChars(newfile)
            with open("C:\\Users\\IBM_ADMIN\\Documents\\1 - Clients\\HTC\\sma_for_import\\" + removeDisallowedFilenameChars(newfile), 'wb') as csvoutput:
                writer = csv.writer(csvoutput, lineterminator='\r\n', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                for m in terms:
                    m = m.encode('ascii', 'ignore')
                    m = m.replace("\"\"","\"")
                    if len(m) > 0:
                        print m
                        writer.writerow([m])
        
    
```
