```python
import itertools
import csv

codefrom = '0319T'
codeto = '0325T'

# I want something like the following:
#for i in range(codefrom, codeto):
#    print i

def next_string(s):
  a1 = range(65, 91)  # capital letters
  a2 = range(97, 123) # letters
  a3 = range(48, 58)  # numbers
  char = ord(s[-1])
  for a in [a1, a2, a3]:
    if char in a:
      if char + 1 in a:
        return s[:-1] + chr(char + 1)
      else:
        ns = next_string(s[:-1]) if s[:-1] else chr(a[0])
        return ns + chr(a[0])

def alphnumeric_range(codefrom, codeto):
    values = []
    values.append(codefrom)
    i = codefrom
    while i < codeto:
        i = next_string(i)
        values.append(i)
    return values
    
#print alphnumeric_range(codefrom, codeto)
with open('C:/Projects/Tenet/Contract Term Lookup/Cigna Health Plan/reimbursement table/CIGNA NATIONAL PACEMAKER.csv', "rb") as input:
    lookup_table = zip(*csv.reader(input))
    lookup_list = []
    for i in range(1,len(lookup_table[0])):
        lookup_list.extend(alphnumeric_range(lookup_table[0][i], lookup_table[1][i]))
    CIGNA_NATIONAL_PACEMAKER = lookup_list  
    
print CIGNA_NATIONAL_PACEMAKER
```
